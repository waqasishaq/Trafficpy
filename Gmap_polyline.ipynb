```python
base_url = "https://maps.googleapis.com/maps/api/directions/json?"
```

```python
# geocoordinates
origin = "New brunswick, nj"
destination ="jacson heights, ny"
api_key = "AIzaSyClUQ2ca81McBm42Z5siXFRUFuDcQZccY4"

# set up a parameters dictionary
params = {
    "origin": origin,
    "destination": destination,
    "key": api_key
}
```

```python
# run a request using our params dictionary
response = requests.get(base_url, params=params)
```

```python
print(response.url)
```

```python
# convert response to json
Direction_data = response.json()

# Print the json (pretty printed)
print(json.dumps(Direction_data, indent=4, sort_keys=True))
```

```python
cordinates=[]
cordinates.append((Direction_data["routes"][0]["legs"][0]["start_location"]["lat"],Direction_data["routes"][0]["legs"][0]["start_location"]["lng"]))

for items in range(len(Direction_data["routes"][0]["legs"][0]["steps"])):
    cordinates.append((Direction_data["routes"][0]["legs"][0]["steps"][items]["end_location"]["lat"],Direction_data["routes"][0]["legs"][0]["steps"][items]["end_location"]["lng"]))
    
```
