```python
# 7200 seconds is 2 hours, so 
# if latest departuretime-nowtime > 7200
    #itterate -5 mins for 24 itterations
    #else: time = latestdeparttime-nowtime, time/300 = #of itterations
```

```python
timelist = []
timechange = timestamp
if timestamp - now > 7200:
    for x in range(0,24):
        timelist.append(timechange)
        timechange += -300
        
else:
    numb = int((timestamp-now)/60/5)
    for x in range(0,numb):
        timelist.append(timechange)
        timechange += -300
        

timelist2 = [time.ctime((x)) for x in timelist]
timelist3 = [datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%dT%H:%M:%S') for x in timelist]
```
